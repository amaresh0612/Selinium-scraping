```python

s =web.get_page_source()
soup = BeautifulSoup(s,"html.parser")
l1=[]
all = soup.find_all("div",{"class":"organisation"})
for item in all:
    d1= {}
    d1["Contact Name"]=item.find("h3").text.replace("\n","").replace("  ","")
    try:
        d1["Bio"]=item.find("div",{"class":"profiletext"}).text.replace("\n",",")
    except:
        d1["Bio"]=None
    l1.append(d1)

l2=[]
all_b =soup.find_all("div",{"class":"attendee-profile"})
for item in all_b:
    d2={}
    try:
        d2["Company Name"]=item.h5.text
    except:
        d2["Company Name"]=None      
    try:
        d2["Title"]=item.em.text
    except:
        d2["Title"]=None

    l2.append(d2)

l3=[] 
all_web= soup.find_all("div",{"class":"organisation-details"})        
for item in all_web:
    d3={}
    try:
        d3["website"]=item.a.text
    except:
        d3["website"]=None
    l3.append(d3)

l4=[]        
all_niche = soup.find_all("div",{"class":"personal-profile"})
for item in all_niche:
    d4={}
    try:
        d4["Niche"]=item.li.a.text
    except:
        d4["Niche"]=None
    l4.append(d4)

df1= pd.DataFrame(l1)
df2= pd.DataFrame(l2)
df3= pd.DataFrame(l3)
df4= pd.DataFrame(l4)
df =pd.concat([df1,df2,df3,df4], axis=1) 
new_order = [2,1,3,0,4,5]
df = df[df.columns[new_order]]
df.head(10)
```
